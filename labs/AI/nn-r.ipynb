```r
library(dplyr)
```

```r
library(DT)
library(neuralnet)
```

```r
# Technical Knowledge Score
TKS <- c(20, 10, 30, 20, 80, 30)
# Communication Skills Score
CSS <- c(90, 20, 40, 50, 50, 80)
# Student Placed
```

```r

Placed <- c(1, 0, 0, 0, 1, 1)
```

```r

# об'єдеаємо дані в набір
df <- data.frame(TKS, CSS, Placed)
```

```r
df
```

```r
net <- neuralnet(Placed ~ TKS + CSS, data=df, hidden=3, 
                act.fct = "logistic",
                # rep = 200,
                linear.output = FALSE)
```

```r
plot(net, rep = "best")
```

```r
TKS.test <- c(30, 40, 85)
CSS.test <- c(85, 50, 40)
test <- data.frame(TKS.test, CSS.test)
```

```r
test
```

```r
Predict <- predict(net, test)
```

```r
Predict
```

```r
round(Predict)
```

```r
net2 <- neuralnet(Placed ~ TKS + CSS, data=df, hidden=3, 
                act.fct = "tanh",
                rep = 200,
                linear.output = FALSE)
```

```r
plot(net2, rep = "best")
```

```r
Predict2 <- predict(net2, test)
```
